```python
# DataFrame con todos los discursos:
df_speeches = pd.read_csv('../data/us_2020_election_speeches.csv')
df_speeches
```

```python
# Realizar la limpieza de los datos que crea pertinente. Se espera que usen la función clean_text() de la entrega anterior.

nlp = spacy.load("es_core_news_sm")

#Agregamos una funcion para eliminar las stop words y lematizar el texto?

def lemmatize_and_remove_stopwords(texto):
    doc = nlp(texto)
    tokens_limpios = [
        token.lemma_ for token in doc
        if not token.is_stop and not token.is_punct and not token.is_space
    ]
    return " ".join(tokens_limpios)

def clean_text(df, column_name):
    # Eliminar primeras palabras hasta el primer "\n"
    result = df[column_name].str.replace(r"^[^\n]*\n", "", regex=True)

    # Convertir todo a minúsculas
    result = result.str.lower()

    # Completar signos de puntuación faltantes
    for punc in ["[", "\n", ",", ":", "?"]:
        result = result.str.replace(punc, " ")

    # Aplicar lematización y eliminación de stopwords a cada texto limpio
    # result = result.apply(lemmatize_and_remove_stopwords)

    return result

#Creamos una nueva columna CleanText a partir de text
df_speeches["CleanText"] = clean_text(df_speeches, 'text')
```
